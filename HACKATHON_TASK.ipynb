```python
# CELL 1: INSTALL LIBRARIES
import sys
!{sys.executable} -m pip install openai pandas numpy --quiet
print("âœ… Libraries installed!")

# CELL 2: COMPLETE CA-GPT WITH FULL FINANCIAL REPORTING
import openai
import os
import time
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from typing import List, Dict, Any, Optional
from dataclasses import dataclass, field
import warnings
warnings.filterwarnings('ignore')

# ðŸ”´ CHANGE THIS LINE WITH YOUR REAL OPENAI API KEY
os.environ['OPENAI_API_KEY'] = 'sk-proj-YourActualApiKeyHere'  # â† CHANGE THIS

client = openai.OpenAI()

@dataclass
class AccountTransaction:
    date: str
    description: str
    debit: float
    credit: float
    balance: float
    account_type: str  # Asset, Liability, Equity, Income, Expense

@dataclass
class Investment:
    type: str  # Stocks, Mutual Funds, FD, PPF, Real Estate, Gold
    purchase_date: str
    purchase_value: float
    current_value: float
    returns_percentage: float
    holding_period: str
    tax_implication: str

@dataclass
class ClientAccount:
    client_name: str
    pan_number: str
    gst_number: str
    financial_year: str
    transactions: List[AccountTransaction] = field(default_factory=list)
    investments: List[Investment] = field(default_factory=list)
    opening_balance: float = 0.0
    closing_balance: float = 0.0
    total_debits: float = 0.0
    total_credits: float = 0.0

class CharteredAccountantSystem:
    """Complete CA reporting system with investments, debits, credits"""
    
    def __init__(self):
        self.clients = {}
    
    def create_sample_account(self) -> ClientAccount:
        """Create a sample client account with transactions and investments"""
        
        account = ClientAccount(
            client_name="TechSolutions Pvt Ltd",
            pan_number="AABCT1234E",
            gst_number="27AAPFU0939F1Z5",
            financial_year="2024-25"
        )
        
        # Add transactions (credits and debits)
        transactions = [
            AccountTransaction("2024-04-01", "Opening Balance", 0, 0, 500000, "Opening"),
            AccountTransaction("2024-04-05", "Sales Revenue - Client A", 0, 150000, 650000, "Income"),
            AccountTransaction("2024-04-10", "Office Rent Payment", 50000, 0, 600000, "Expense"),
            AccountTransaction("2024-04-15", "IT Services - Client B", 0, 200000, 800000, "Income"),
            AccountTransaction("2024-04-20", "Salary Payment", 120000, 0, 680000, "Expense"),
            AccountTransaction("2024-04-25", "Electricity Bill", 15000, 0, 665000, "Expense"),
            AccountTransaction("2024-05-01", "Consulting Fees - Client C", 0, 180000, 845000, "Income"),
            AccountTransaction("2024-05-05", "Software Purchase", 75000, 0, 770000, "Asset"),
            AccountTransaction("2024-05-10", "GST Payment", 32400, 0, 737600, "Liability"),
            AccountTransaction("2024-05-15", "Dividend Received", 0, 25000, 762600, "Income"),
            AccountTransaction("2024-05-20", "Marketing Expenses", 35000, 0, 727600, "Expense"),
            AccountTransaction("2024-05-25", "Professional Fees", 0, 95000, 822600, "Income"),
            AccountTransaction("2024-06-01", "TDS Deducted", 18000, 0, 804600, "Liability"),
            AccountTransaction("2024-06-05", "Bank Interest", 0, 12000, 816600, "Income"),
            AccountTransaction("2024-06-10", "Insurance Premium", 25000, 0, 791600, "Expense"),
            AccountTransaction("2024-06-15", "Equipment Purchase", 100000, 0, 691600, "Asset"),
        ]
        
        # Calculate running balance and totals
        balance = account.opening_balance
        total_debits = 0
        total_credits = 0
        
        for t in transactions:
            balance = balance - t.debit + t.credit
            t.balance = balance
            total_debits += t.debit
            total_credits += t.credit
            account.transactions.append(t)
        
        account.total_debits = total_debits
        account.total_credits = total_credits
        account.closing_balance = balance
        
        # Add investments
        investments = [
            Investment("Fixed Deposit", "2023-04-01", 500000, 537500, 7.5, "1 year", "Taxable as per slab"),
            Investment("PPF", "2022-06-01", 150000, 171000, 14.0, "2 years", "EEE - Exempt"),
            Investment("Mutual Funds", "2023-01-15", 200000, 248000, 24.0, "1.5 years", "LTCG > 1L taxed at 10%"),
            Investment("Stocks - Reliance", "2024-02-01", 100000, 112000, 12.0, "4 months", "STCG taxed at 15%"),
            Investment("Gold ETF", "2023-11-01", 75000, 81750, 9.0, "7 months", "LTCG if held >3 years"),
            Investment("Corporate Bonds", "2024-01-01", 300000, 309000, 3.0, "5 months", "Taxable as per slab"),
        ]
        
        account.investments = investments
        return account
    
    def generate_ledger_report(self, account: ClientAccount) -> str:
        """Generate detailed ledger with debits and credits"""
        
        report = f"""
{'='*80}
                    CHARTERED ACCOUNTANT - LEDGER REPORT
{'='*80}

CLIENT DETAILS:
-------------
Name: {account.client_name}
PAN: {account.pan_number}
GSTIN: {account.gst_number}
Financial Year: {account.financial_year}
Report Date: {datetime.now().strftime('%d-%m-%Y %H:%M')}

{'='*80}
                    GENERAL LEDGER - ALL TRANSACTIONS
{'='*80}

Date       | Description                  | Debit (Rs.)   | Credit (Rs.)  | Balance (Rs.) 
-----------|------------------------------|-------------|-------------|-------------
"""
        for t in account.transactions:
            report += f"{t.date} | {t.description:<28} | {t.debit:>10,.2f} | {t.credit:>10,.2f} | {t.balance:>11,.2f}\n"
        
        report += f"""
{'='*80}
                    SUMMARY OF TRANSACTIONS
{'='*80}

Total Debits:  Rs. {account.total_debits:>15,.2f}
Total Credits: Rs. {account.total_credits:>15,.2f}
Opening Balance: Rs. {account.opening_balance:>12,.2f}
Closing Balance: Rs. {account.closing_balance:>12,.2f}

Net Profit/Loss: Rs. {account.total_credits - account.total_debits:>13,.2f}

{'='*80}
"""
        return report
    
    def generate_investment_report(self, account: ClientAccount) -> str:
        """Generate detailed investment portfolio report"""
        
        report = f"""
{'='*80}
                    INVESTMENT PORTFOLIO REPORT
{'='*80}

INVESTMENT SUMMARY:
------------------
Total Client: {account.client_name}
Valuation Date: {datetime.now().strftime('%d-%m-%Y')}

{'='*80}
Investment Type   | Purchase Date | Purchase (Rs.) | Current (Rs.) | Returns % | Period    | Tax Implication
------------------|---------------|--------------|-------------|-----------|-----------|-----------------
"""
        total_purchase = 0
        total_current = 0
        
        for inv in account.investments:
            report += f"{inv.type:<16} | {inv.purchase_date} | {inv.purchase_value:>11,.2f} | {inv.current_value:>11,.2f} | {inv.returns_percentage:>7.1f}% | {inv.holding_period:<9} | {inv.tax_implication:<15}\n"
            total_purchase += inv.purchase_value
            total_current += inv.current_value
        
        total_returns = total_current - total_purchase
        returns_percentage = (total_returns / total_purchase * 100) if total_purchase > 0 else 0
        
        report += f"""
{'='*80}
PORTFOLIO SUMMARY:
-----------------
Total Investment Value (Purchase): Rs. {total_purchase:>15,.2f}
Total Investment Value (Current):  Rs. {total_current:>15,.2f}
Absolute Returns:                  Rs. {total_returns:>15,.2f}
Returns Percentage:                {returns_percentage:>14.2f}%

{'='*80}
TAX PLANNING RECOMMENDATIONS:
---------------------------
1. Section 80C: Invest up to â‚¹1,50,000 in PPF/ELSS for tax deduction
2. Long-term Capital Gains: Hold equity investments for >1 year for LTCG benefit
3. Debt Funds: Hold for >3 years for indexation benefit
4. Tax-free bonds: Consider for tax-free interest income

{'='*80}
"""
        return report
    
    def generate_profit_loss_report(self, account: ClientAccount) -> str:
        """Generate Profit & Loss Statement"""
        
        # Categorize transactions
        income = sum(t.credit for t in account.transactions if t.account_type == "Income")
        expenses = sum(t.debit for t in account.transactions if t.account_type == "Expense")
        gross_profit = income - expenses
        
        report = f"""
{'='*80}
                    PROFIT & LOSS STATEMENT
{'='*80}

For the Period: {account.financial_year}
Company: {account.client_name}

{'='*80}
INCOME:
{'='*80}
Sales Revenue                           Rs. {income:>35,.2f}
Other Income                           Rs. {0:>35,.2f}
{'â”€'*80}
TOTAL INCOME                           Rs. {income:>35,.2f}

{'='*80}
EXPENSES:
{'='*80}
Operating Expenses                     Rs. {expenses:>35,.2f}
Administrative Expenses               Rs. {0:>35,.2f}
Depreciation                          Rs. {0:>35,.2f}
{'â”€'*80}
TOTAL EXPENSES                        Rs. {expenses:>35,.2f}

{'='*80}
PROFIT BEFORE TAX                     Rs. {gross_profit:>35,.2f}

Tax Estimate (25.17%):                Rs. {gross_profit * 0.2517:>35,.2f}
Profit After Tax:                     Rs. {gross_profit * 0.7483:>35,.2f}

{'='*80}
"""
        return report
    
    def generate_balance_sheet(self, account: ClientAccount) -> str:
        """Generate Balance Sheet"""
        
        # Calculate assets and liabilities
        cash_balance = account.closing_balance
        investments_total = sum(inv.current_value for inv in account.investments)
        fixed_assets = sum(t.debit for t in account.transactions if t.account_type == "Asset")
        
        total_assets = cash_balance + investments_total + fixed_assets
        total_liabilities = sum(t.debit for t in account.transactions if t.account_type == "Liability")
        capital = total_assets - total_liabilities
        
        report = f"""
{'='*80}
                    BALANCE SHEET
{'='*80}

As on: {datetime.now().strftime('%d-%m-%Y')}
Company: {account.client_name}

{'='*80}
ASSETS                           Amount (Rs.)      LIABILITIES                    Amount (Rs.)
{'='*80}

Current Assets:                              Current Liabilities:
Cash & Bank            {cash_balance:>14,.2f}      GST Payable               {total_liabilities:>16,.2f}
Investments           {investments_total:>14,.2f}      TDS Payable                      0.00
Fixed Assets          {fixed_assets:>14,.2f}      Other Liabilities                 0.00
Other Assets                    0.00      
                                           â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€                         â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
TOTAL ASSETS          {total_assets:>14,.2f}      TOTAL LIABILITIES       {total_liabilities:>16,.2f}

                                           SHAREHOLDER'S EQUITY
                                           Share Capital                    500,000.00
                                           Retained Earnings          {capital - 500000:>14,.2f}
                                           â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
                                           TOTAL EQUITY              {capital:>16,.2f}

{'='*80}
TOTAL LIABILITIES & EQUITY      {total_liabilities + capital:>14,.2f}
{'='*80}
"""
        return report
    
    def generate_complete_ca_report(self, account: ClientAccount) -> str:
        """Generate complete Chartered Accountant report with all details"""
        
        report = f"""
{'='*100}
                    CHARTERED ACCOUNTANT - COMPREHENSIVE FINANCIAL REPORT
{'='*100}

{'='*100}
                         {account.client_name}
                         PAN: {account.pan_number}
                         GST: {account.gst_number}
                         Financial Year: {account.financial_year}
{'='*100}

"""
        report += self.generate_ledger_report(account)
        report += self.generate_profit_loss_report(account)
        report += self.generate_balance_sheet(account)
        report += self.generate_investment_report(account)
        
        # Add CA Recommendations
        report += f"""
{'='*100}
                    CHARTERED ACCOUNTANT - AUDIT OBSERVATIONS & RECOMMENDATIONS
{'='*100}

1. TAX COMPLIANCE:
   âœ“ GST Returns filed regularly
   âœ“ TDS deducted and deposited on time
   âš  Advance tax payments due for next quarter

2. AUDIT FINDINGS:
   âœ“ Books of accounts are properly maintained
   âœ“ All transactions are supported by valid documents
   âš  Inventory valuation method needs consistency

3. INVESTMENT ADVICE:
   âœ“ Consider diversifying into debt funds for stability
   âš  Review underperforming mutual funds
   âœ“ PPF extension recommended for continued tax benefits

4. RISK ASSESSMENT:
   â€¢ Credit Risk: Low - Healthy collection period
   â€¢ Market Risk: Medium - Equity exposure in portfolio
   â€¢ Compliance Risk: Low - Regular filings maintained

5. RECOMMENDATIONS FOR NEXT QUARTER:
   â€¢ Increase Section 80C investments before March
   â€¢ Reconcile GST input credit claims
   â€¢ Review outstanding receivables over 90 days
   â€¢ Consider health insurance for additional 80D deduction

{'='*100}
Report Generated by: CA-GPT AI System
Chartered Accountant (Virtual)
Date: {datetime.now().strftime('%d-%m-%Y %H:%M:%S')}
{'='*100}

DISCLAIMER: This report is AI-generated and should be reviewed 
by a qualified Chartered Accountant before finalization.

{'='*100}
"""
        return report

# Initialize the CA system
ca_system = CharteredAccountantSystem()

# Create sample account with complete data
sample_account = ca_system.create_sample_account()

# Generate complete CA report
complete_report = ca_system.generate_complete_ca_report(sample_account)

# Display in notebook
print(complete_report)

# Save with UTF-8 encoding
with open('chartered_accountant_full_report.txt', 'w', encoding='utf-8') as f:
    f.write(complete_report)
print("âœ… Complete CA Report saved as 'chartered_accountant_full_report.txt'")

# Also save as CSV for transactions
import csv

with open('transactions_ledger.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Date', 'Description', 'Debit (Rs.)', 'Credit (Rs.)', 'Balance (Rs.)', 'Type'])
    for t in sample_account.transactions:
        writer.writerow([t.date, t.description, t.debit, t.credit, t.balance, t.account_type])
print("âœ… Transaction ledger saved as 'transactions_ledger.csv'")

# Investment portfolio CSV
with open('investment_portfolio.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Type', 'Purchase Date', 'Purchase Value', 'Current Value', 'Returns %', 'Holding Period', 'Tax Implication'])
    for inv in sample_account.investments:
        writer.writerow([inv.type, inv.purchase_date, inv.purchase_value, inv.current_value, 
                        inv.returns_percentage, inv.holding_period, inv.tax_implication])
print("âœ… Investment portfolio saved as 'investment_portfolio.csv'")

# Summary Dashboard
print(f"""
{'='*60}
                    FINANCIAL DASHBOARD SUMMARY
{'='*60}

CLIENT: {sample_account.client_name}

ðŸ’° CASH POSITION:
   Opening Balance: Rs. {sample_account.opening_balance:,.2f}
   Closing Balance: Rs. {sample_account.closing_balance:,.2f}
   Total Debits:    Rs. {sample_account.total_debits:,.2f}
   Total Credits:   Rs. {sample_account.total_credits:,.2f}

ðŸ“Š INVESTMENT PORTFOLIO:
   Total Invested:  Rs. {sum(i.purchase_value for i in sample_account.investments):,.2f}
   Current Value:   Rs. {sum(i.current_value for i in sample_account.investments):,.2f}
   Total Returns:   Rs. {sum(i.current_value for i in sample_account.investments) - sum(i.purchase_value for i in sample_account.investments):,.2f}

ðŸ“ˆ PROFITABILITY:
   Total Income:    Rs. {sum(t.credit for t in sample_account.transactions if t.account_type == 'Income'):,.2f}
   Total Expenses:  Rs. {sum(t.debit for t in sample_account.transactions if t.account_type == 'Expense'):,.2f}
   Net Profit:      Rs. {sum(t.credit for t in sample_account.transactions if t.account_type == 'Income') - sum(t.debit for t in sample_account.transactions if t.account_type == 'Expense'):,.2f}

âœ… REPORTS GENERATED:
   1. chartered_accountant_full_report.txt - Complete CA report
   2. transactions_ledger.csv - All debits & credits
   3. investment_portfolio.csv - Investment details

{'='*60}
""")
```
